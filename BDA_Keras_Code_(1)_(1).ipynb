***Simple Neural Network with Keras Sequential API***

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import numpy as np

# Generate some random data
x_train = np.random.random((1000, 20))  # 1000 samples, 20 features each
y_train = np.random.randint(2, size=(1000, 1))  # Binary labels (0 or 1)

x_test = np.random.random((200, 20))  # 200 test samples
y_test = np.random.randint(2, size=(200, 1))  # Binary labels for testing

# Build a Sequential model
model = Sequential()

# Add a hidden layer with 64 neurons and ReLU activation
model.add(Dense(64, activation='relu', input_shape=(20,)))

# Add another hidden layer with 32 neurons and ReLU activation
model.add(Dense(32, activation='relu'))

# Add an output layer with 1 neuron and sigmoid activation for binary classification
model.add(Dense(1, activation='sigmoid'))

# Compile the model
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

# Train the model
model.fit(x_train, y_train, epochs=5, batch_size=32)

# Evaluate the model on the test data
test_loss, test_acc = model.evaluate(x_test, y_test)
print(f'Test accuracy: {test_acc}')
```

```python
# ✅ Step 1: Import Libraries
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.datasets import mnist
from tensorflow.keras.utils import to_categorical
import matplotlib.pyplot as plt  # For plotting accuracy

# ✅ Step 2: Load and Preprocess MNIST Dataset
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Normalize pixel values
x_train = x_train.astype('float32') / 255.0
x_test = x_test.astype('float32') / 255.0

# One-hot encode labels
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)

# ✅ Step 3: Build the Model
model = Sequential()

# Flatten input (28x28 -> 784)
model.add(Flatten(input_shape=(28, 28)))

# ✅ Include 5 Hidden Layers with ReLU
model.add(Dense(256, activation='relu'))   # Layer 1
model.add(Dense(128, activation='relu'))   # Layer 2
model.add(Dense(64, activation='relu'))    # Layer 3
model.add(Dense(32, activation='relu'))    # Layer 4
model.add(Dense(16, activation='relu'))    # Layer 5

# ✅ Output Layer with Softmax for 10-class classification
model.add(Dense(10, activation='softmax'))

# ✅ Step 4: Compile the Model with Adamax optimizer
model.compile(
    optimizer='adamax',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# ✅ Step 5: Train the Model for 20 Epochs and capture history
history = model.fit(
    x_train, y_train,
    epochs=30,
    batch_size=32,
    validation_split=0.2,
    verbose=1
)

# ✅ Step 6: Evaluate Model on Test Data
test_loss, test_acc = model.evaluate(x_test, y_test)
print(f"\n✅ Final Test Accuracy: {test_acc * 100:.2f}%")

# ✅ Step 7: Plot Training vs Validation Accuracy
plt.plot(history.history['accuracy'], label='Train Accuracy', marker='o')
plt.plot(history.history['val_accuracy'], label='Val Accuracy', marker='x')
plt.title('Training vs Validation Accuracy (ReLU + Adamax)')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Dropout
from tensorflow.keras.datasets import mnist
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.callbacks import EarlyStopping

# Load and preprocess MNIST dataset
(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train = x_train.astype('float32') / 255.0
x_test = x_test.astype('float32') / 255.0
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)

# Early stopping callback (monitor val_accuracy, wait 3 epochs, restore best weights)
early_stop = EarlyStopping(monitor='val_accuracy', patience=3, restore_best_weights=True)

# Function to build and train the model
def build_and_train_model(activation_fn, optimizer_name):
    model = Sequential()
    model.add(Flatten(input_shape=(28, 28)))

    # Add more layers with dropout to prevent overfitting
    model.add(Dense(512, activation=activation_fn))
    model.add(Dropout(0.3))  # Drop 30% neurons
    model.add(Dense(256, activation=activation_fn))
    model.add(Dropout(0.3))
    model.add(Dense(128, activation=activation_fn))
    model.add(Dropout(0.2))
    model.add(Dense(64, activation=activation_fn))
    model.add(Dense(32, activation=activation_fn))

    # Output layer
    model.add(Dense(10, activation='softmax'))

    # Compile
    model.compile(
        optimizer=optimizer_name,
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )

    print(f"\nTraining with activation='{activation_fn}' and optimizer='{optimizer_name}'")

    # Train the model with early stopping
    model.fit(
        x_train, y_train,
        epochs=25,
        batch_size=32,
        validation_split=0.2,
        verbose=1,
        callbacks=[early_stop]
    )

    # Evaluate
    loss, acc = model.evaluate(x_test, y_test, verbose=0)
    print(f"✅ Test Accuracy: {acc * 100:.2f}%\n")
    return acc * 100

# Define configurations
activation_functions = ['relu']
optimizers = ['adamax']

# Run all combinations
results = {}
for act in activation_functions:
    for opt in optimizers:
        acc = build_and_train_model(act, opt)
        results[(act, opt)] = acc

# Display results in a readable format
print("📊 Summary of Results:\n")
for key, val in results.items():
    print(f"Activation: {key[0]:7s} | Optimizer: {key[1]:7s} | Accuracy: {val:.2f}%")
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Dropout, BatchNormalization
from tensorflow.keras.datasets import mnist
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.callbacks import EarlyStopping

# ✅ Step 1: Load and preprocess MNIST dataset
(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train = x_train.astype('float32') / 255.0
x_test = x_test.astype('float32') / 255.0
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)

# ✅ Step 2: EarlyStopping callback to prevent overfitting
early_stop = EarlyStopping(
    monitor='val_accuracy',
    patience=5,
    restore_best_weights=True
)

# ✅ Step 3: Build the optimized model
model = Sequential()
model.add(Flatten(input_shape=(28, 28)))  # Input layer

# Hidden Layers with more neurons, dropout & batch normalization
model.add(Dense(1024, activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.3))

model.add(Dense(512, activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.3))

model.add(Dense(256, activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.2))

# Output Layer
model.add(Dense(10, activation='softmax'))

# ✅ Step 4: Compile the model
model.compile(
    optimizer='adamax',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# ✅ Step 5: Train the model
history = model.fit(
    x_train, y_train,
    epochs=50,                  # Will likely stop earlier via EarlyStopping
    batch_size=16,              # Smaller batch size for better updates
    validation_split=0.2,
    callbacks=[early_stop],
    verbose=1
)

# ✅ Step 6: Evaluate model
loss, acc = model.evaluate(x_test, y_test, verbose=0)
print(f"\n🎯 Final Test Accuracy: {acc * 100:.2f}%")

# ✅ Step 7: Plot accuracy
import matplotlib.pyplot as plt

plt.plot(history.history['accuracy'], label='Train Accuracy', marker='o')
plt.plot(history.history['val_accuracy'], label='Val Accuracy', marker='x')
plt.title('Training vs Validation Accuracy (Optimized Model)')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)
plt.show()
```
